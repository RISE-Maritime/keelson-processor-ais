```python
# NMEA Parser
import pandas as pd
import pynmea2
import pyais
```

```python
df_org = pd.read_json('nmea.json')
df_org
```

```python
df = df_org[['topic']].join(df_org['data'].apply(pd.Series))
df['timestamp'] = pd.to_datetime(df['timestamp'])
df
```

```python
df['first_segment'] = df['value'].str.split(',').str[0]
df.first_segment.unique()
```

```python
import geopy.distance
```

```python

lat = 60.00
long = 10.00
heading = 358
to_bow = 10
to_stern = 40
to_starboard = 5
to_port = 15

# Longitudinal
midpoint = (to_bow + to_stern) / 2
if midpoint > to_bow:
    move_long = to_bow - midpoint
elif midpoint < to_bow:
    move_long = to_bow - midpoint
else:
    move_long = 0

# Lateral
midpoint = (to_starboard + to_port) / 2
if midpoint > to_starboard:
    move_lat = to_starboard - midpoint
elif midpoint < to_starboard:
    move_lat = to_starboard - midpoint
else:
    move_lat = 0

point_long_adj = geopy.distance.distance(meters=move_long).destination((lat, long), bearing=heading)
point_final = geopy.distance.distance(meters=move_lat).destination((point_long_adj.latitude, point_long_adj.longitude), bearing=heading+90)

point_final.latitude, point_final.longitude 

```
